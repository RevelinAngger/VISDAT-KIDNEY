#Func Trend Analysis

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
```

#Tren Analysis saham Hangseng

```python
df_hangseng['Trend']= np.zeros(df_hangseng['Day_Perc_Change'].count())
df_hangseng['Trend']= df_hangseng['Day_Perc_Change'].apply(lambda x:trend(x))
df_hangseng
```

```python
# type your code here
combined_df = pd.concat([df_nikkei["Adj Close"], df_nasdaq["Adj Close"], df_hangseng["Adj Close"]], axis=1)
combined_df.columns = ["df_nikkei", "df_nasdaq", "df_hangseng"]
combined_df.head()
```

```python
combined_df.isnull().sum()
#melakukan pengecekan data null
```

```python
combined_df.dropna(inplace = True, axis = 0)
combined_df.isnull().sum()
#menghilangkan  data kosong dari ketiga saham
```

```python
pct_chg_df = combined_df.pct_change()*100
pct_chg_df
#presentase perubahan saham
```

#Volati Saham Dari Hangseng

```python
df_hangseng = pct_chg_df["df_hangseng"].rolling(7).std()*np.sqrt(7)
print('Volatilitas Saham Hangseng')
df_hangseng.plot(figsize = (15, 7))
```

#Volati Saham Dari Nasdaq

```python
df_nasdaq = pct_chg_df["df_nasdaq"].rolling(7).std()*np.sqrt(7)
print('Volatilitas Saham Nasdaq')
df_nasdaq.plot(figsize = (15, 7))
```

#Volati Saham Dari Nikkei

```python
df_nikkei = pct_chg_df["df_nikkei"].rolling(7).std()*np.sqrt(7)
print('Volatilitas Saham Nikkei')
df_nikkei.plot(figsize = (15, 7))
```
